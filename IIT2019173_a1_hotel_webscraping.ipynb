```python

def data_covid():
    URL="https://www.worldometers.info/coronavirus/#countries"
    html_page=requests.get(URL).text
    soup=BeautifulSoup(html_page,'lxml')


    get_table=soup.find("table",id="main_table_countries_today")
    get_table_data=get_table.tbody.find_all("tr")
    # get_table_data.td.find_all("a")
    # get_table_data


    dic={}
    for tr in get_table_data:
        try:
            key=tr.find("a",href=True).string
        except:
            key=tr.find("td").string
        values=[j.string for j in tr.find_all('td')]
        dic[key]=values
        #print(key)
    column_names=["Total Cases","New Cases","Total Deaths","New Deaths","Total Recovered","New Recovered","Active","Serious/Critical","Cases/million","Deaths/million","Total Tests","Tests/million"]
    df=pd.DataFrame(dic).iloc[1:,:].T.iloc[1:,1:13]
    df.index_name="Country"
    df.columns=column_names
    df.to_csv("Covid_data.csv")
```
