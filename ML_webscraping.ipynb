```python
import numpy as np
import pandas as pd
import urllib
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.goibibo.com/hotels-international/hotels-in-langkawi-ct/"
headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }
response = requests.request("GET", url, headers=headers)
data = BeautifulSoup(response.text, 'html.parser')
print(data)
```

```python
cards_data = data.find_all('div', attrs={'class', 'HotelCardstyles__OuterWrapperDiv-sc-1s80tyk-0 cvGVYW'})
print('Total Number of Cards Found : ', len(cards_data))
#for card in cards_data:
#print(card)
```

```python
i = 1
for card in cards_data:
  hotel_name = card.find('a',attrs={'itemprop':'name'})
  url_to_visit = 'https://www.goibibo.com/' + hotel_name['href']
  card_response = requests.request("GET", url_to_visit, headers=headers)
  detailed_data = BeautifulSoup(card_response.text, 'html.parser')
  description = detailed_data.find('div',attrs={'class':'AboutPropertiesstyles__MainContentTextStyle-sc-4xqpe7-2 knLjJo'})
  city_outer = card.find('div',attrs={'class':'PersuasionHoverTextstyles__TextWrapperSpan-sc-1c06rw1-14 eMOIOB'})
  if(city_outer):
    city = city_outer.find('span')
  star_rating_outer = card.find('span',attrs={'itemprop':'starRating'})
  price = card.find('p',attrs={'class':'HotelCardstyles__CurrentPrice-sc-1s80tyk-28 inUyrJ'})
  amenities = card.find_all('p',attrs={'class':'AmenitiesListstyles__TextWrapper-sc-19dqtu1-7 iEiBKc'})
  if(city_outer):
    if star_rating_outer:
      print(i,hotel_name.text,city.text,star_rating_outer['content'],price.text)
    else:
      print(i,hotel_name.text,city.text,price.text)
  else:
    print(i,hotel_name.text,star_rating_outer['content'],price.text)
  print(description.text)
  i = i+1
  j = 1
  for amenity in amenities:
    print(j,amenity.text)
    j = j+1
  print("----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
```
