# Q1

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
listx = [2006,2008,2009,2011,2013,2014,2015,2016,2017,2018,2019]
listy = [100.2,98.3,87.1,89.2,88.9,83.5,89.1,84,92.3,96,97]
m=len(listx)
year=np.array(listx)
revenue=np.array(listy)
meanx=np.mean(year)
meany=np.mean(revenue)
meanxlist=np.ones([m], dtype=np.uint) *meanx
meanylist=np.ones([m], dtype=np.uint) *meany
devx=np.subtract(year,meanxlist)
devy=np.subtract(revenue,meanylist)
devprod=np.multiply(devx,devy)
devx2=np.square(devx)
slope=sum(devprod)/sum(devx2)
intercept=meany-slope*meanx
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(year,revenue)
x = np.linspace(2005,2021,100)
y = intercept+x*slope
ax1.plot(x, y, '-r')
plt.title('Least Square Line Fit')
plt.xlabel('Year')
plt.ylabel('Revenue')
print("Expected revenue in 2021:",intercept+2021*slope)
errorsum=0
for i in range(m):
    errorsum+=pow(intercept+year[i]*slope-revenue[i],2)
print("Expected error:",errorsum/(2*m))
```

# Solution
P*V^n=C <br/>
ln(P)+ n*ln(V)= ln(C) <br/>
ln(P) = ln(C) + n*ln(1/V) <br/>
This is in the form of a linear equation. Therefore, Y is ln(P), X is ln(1/V). <br/>

```python
import math
listx = [54.3,61.8,72.4,88.7,118.6,194]
listy = [61.2,49.5,37.5,28.4,19.2,10.1]
m=len(listx)
V=np.array(listx)
P=np.array(listy)
vinv= np.reciprocal(V.astype(float))
lnv=np.log(vinv)
lnp=np.log(P)
meanx=np.mean(lnv)
meany=np.mean(lnp)
meanxlist=np.ones([m], dtype=np.uint) *meanx
meanylist=np.ones([m], dtype=np.uint) *meany
devx=np.subtract(lnv,meanxlist)
devy=np.subtract(lnp,meanylist)
devprod=np.multiply(devx,devy)
devx2=np.square(devx)
slope=sum(devprod)/sum(devx2)
intercept=meany-slope*meanx
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(V,P)
x = np.linspace(1,200,100)
y = math.exp(intercept)/pow(x,slope)
ax1.plot(x, y, '-r')
plt.title('P vs V')
plt.xlabel('Volume')
plt.ylabel('Pressure')
print("Value of C:",math.exp(intercept))
print("Value of n:",slope)
print("Equation connecting P and V is P*V^",slope,"=",math.exp(intercept))
print("Estimated Pressure when Volume=100:",intercept+100*slope)
intercept
```
